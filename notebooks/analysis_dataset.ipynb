```python
import os
import glob
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_j = pd.read_csv("../DMP_9D_ACCEL_Logs/TEST/curl_4_22_curl3.csv")
print(len(df_j))
```

```python
df_o = pd.read_csv("../DMP_9D_ACCEL_Logs/TEST/complex555.csv")
print(len(df_o))
```

```python
tSec_j = (df_j['timeMs'] - df_j['timeMs'].iloc[0])/1000
```

```python
tSec_o = (df_o['timeMs'] - df_o['timeMs'].iloc[0])/1000
```

```python
for i in range(0, len(tSec_o), 500):
    print(tSec_o[i])
```

```python
plt.figure(figsize=[10, 5])
plt.plot(tSec_j, df_j['ax'], label='ax')
plt.plot(tSec_j, df_j['ay'], label='ay')
plt.plot(tSec_j, df_j['az'], label='az')
for i in range(0, len(tSec_j), 500):
    print(tSec_j[i])
    plt.vlines(tSec_j[i], -15000, 15000)
plt.grid()
```
